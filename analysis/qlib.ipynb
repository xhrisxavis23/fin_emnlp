```python
import qlib
from qlib.data import D

qlib.init(provider_uri="~/.qlib/qlib_data/sh_sp500_qlib")  # 본인 경로로

# 예: CSI300 구성종목의 일봉 close, volume
instruments = D.instruments("all")   # 또는 "all"
fields = ["$close", "$open", "$high", "$low", "$volume", "$factor", "$change", '$adjclose']

df = D.features(instruments, fields, freq="day")
print(df.tail())
```
